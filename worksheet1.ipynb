```python
import numpy as np
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A·A⁻¹ = I
A_inv = np.linalg.inv(A) # linalg.inv bhaneko inverse
print("A·A⁻¹ =")
print(A @ A_inv) # @ bhannale multiply

# 2. Prove AB ≠ BA
AB = A @ B
BA = B @ A
print("\nAB =", AB)
print("BA =", BA)

# 3. Prove (AB)ᵀ = BᵀAᵀ
AB_T = (A @ B).T # t bhanale transpose
BT_AT = B.T @ A.T
print("\n(AB)ᵀ =", AB_T)
print("BᵀAᵀ =", BT_AT)

# Solve system: AX = B
coeff = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]])
const = np.array([-1, -3, 9])

# Using inverse method: X = A⁻¹B
solution = np.linalg.inv(coeff) @ const
print("\nSolution:", solution)

# Using np.linalg.solve (more efficient)
solution2 = np.linalg.solve(coeff, const)
print("Solution (linalg.solve):", solution2)
```
