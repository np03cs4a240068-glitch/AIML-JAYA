```python


import pandas as pd

# dataset load hareko

data = pd.read_csv("titanic.csv")
print("DATASET LOADED")

# first rows display
print("\nFirst 5 rows of the dataset:")
print(data.head())

# dataframe ko info
print("\nDataset Information:")
print(data.info())



print("\nPROBLEM 1 - SORTING")


# fare coumn
print("\n1. DataFrame with only Fare column:")
fare = data[['Fare']]
print(fare.head())

# pclass ra age ko columns
print("\n2. DataFrame with Pclass and Age columns:")

class_age = data[['Pclass', 'Age']]
print(class_age.head())

# dataframe with survived manxe ra sex columns
print("\n3. DataFrame with Survived and Sex columns:")
survived_gender = data[['Survived', 'Sex']]
print(survived_gender.head())
```

```python


import pandas as pd

print("PROBLEM 2 - SUBSETTING")

print("\n PART 1: SUBSETTING ROWS")


# fare > 100
print("\n1. Passengers with Fare > 100:")
fare_gt_100 = data[data['Fare'] > 100]
print(fare_gt_100.head())
print(f"\nTotal passengers with fare > 100: {len(fare_gt_100)}")

# Pclass filter agreko
print("\n2. First Class Passengers (Pclass = 1):")

first_class = data[data['Pclass'] == 1]
print(first_class.head())
print(f"\nTotal first class passengers: {len(first_class)}")

# filter for female ra age greater than 18
print("\n3. Female passengers under 18:")
female_under_18 = data[(data['Age'] < 18) & (data['Sex'] == 'female')]
print(female_under_18.head())
print(f"\nTotal female passengers under 18: {len(female_under_18)}")

print("\n PART 2: SUBSETTING BY CATEGORICAL VARIABLES")


# filter embarked c ra s
print("\n1. Passengers embarked from Cherbourg (C) or Southampton (S):")
embarked_c_or_s = data[data['Embarked'].isin(['C', 'S'])]
print(embarked_c_or_s.head())
print(f"\nTotal passengers embarked from C or S: {len(embarked_c_or_s)}")

# pclass 1 ra 2
print("\n2. First or Second Class Passengers:")

first_second_class = data[data['Pclass'].isin([1, 2])]
print(first_second_class.head())
print(f"\nTotal first or second class passengers: {len(first_second_class)}")
```

```python


import pandas as pd
import numpy as np


print("SECTION 3.2 - EXPLORATORY DATA ANALYSIS PRACTICE EXERCISE 1")

print("\n QUESTION 1: Highest Fare Relative to Age")


# missing value handle gareko
print("\nHandling missing values in Age column...")
median_age = data['Age'].median()
data['Age'] = data['Age'].fillna(median_age)
print(f"Filled missing ages with median: {median_age}")

# add fare
print("\nStep 1: Adding fare_per_year column (Fare / Age)")
data['fare_per_year'] = data['Fare'] / data['Age']
print(data[['Name', 'Fare', 'Age', 'fare_per_year']].head())

# rows where dare > 5
print("\nStep 2: Filtering for fare_per_year > 5")
high_fare_age = data[data['fare_per_year'] > 5]
print(f"Number of passengers with fare_per_year > 5: {len(high_fare_age)}")

# descending order ma fare sort gareko
print("\nStep 3: Sorting by fare_per_year (descending)")
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

# Name ra fare sort
print("\nStep 4: Selecting Name and fare_per_year columns")
result = high_fare_age_srt[['Name', 'fare_per_year']]

#  Look at the result
print("\nStep 5: Result - Top passengers with highest fare relative to age:")
print(result.head(10))

print("\n ANSWER:")
print(f"Passenger with highest fare relative to age: {result.iloc[0]['Name']}")
print(f"Fare per year: ${result.iloc[0]['fare_per_year']:.2f}")

print("\n QUESTION 2: Adult Male with Highest Fare Relative to Class")


# fare add gareko per class
print("\nStep 1: Adding fare_per_class column (Fare / Pclass)")
data['fare_per_class'] = data['Fare'] / data['Pclass']
print(data[['Name', 'Fare', 'Pclass', 'fare_per_class']].head())

# male ra age >= 18
print("\nStep 2: Filtering for adult males (Age >= 18 and Sex = 'male')")
adult_males = data[(data['Sex'] == 'male') & (data['Age'] >= 18)]
print(f"Number of adult male passengers: {len(adult_males)}")

# descending order ma fare sortng
print("\nStep 3: Sorting by fare_per_class (descending)")
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

# name age ra fare
print("\nStep 4: Selecting Name, Age, and fare_per_class columns")
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]

print("\nStep 5: Result - Top adult male passengers with highest fare relative to class:")
print(result.head(10))

print("\n ANSWER:")
print(f"Adult male passenger with highest fare relative to class: {result.iloc[0]['Name']}")
print(f"Age: {result.iloc[0]['Age']:.0f}")
print(f"Fare per class: ${result.iloc[0]['fare_per_class']:.2f}")
```

```python

import pandas as pd
import numpy as np
print("SECTION 3.3 - EDA WITH GROUP-BY METHOD")

print("\n QUESTION 1: Fare Revenue Percentage by Passenger Class")


# total fare ganeko
print("\nStep 1: Calculate total fare across all passengers")
total_fare = data['Fare'].sum()
print(f"Total fare revenue: ${total_fare:.2f}")

# fare 1class ko lagi
print("\nStep 2: Calculate fare for first class")
first_class_fare = data[data['Pclass'] == 1]['Fare'].sum()
print(f"First class total fare: ${first_class_fare:.2f}")

# second class ko fare
print("\nStep 3: Calculate fare for second class")
second_class_fare = data[data['Pclass'] == 2]['Fare'].sum()
print(f"Second class total fare: ${second_class_fare:.2f}")

# class 3 ko fare
print("\nStep 3: Calculate fare for third class")
third_class_fare = data[data['Pclass'] == 3]['Fare'].sum()
print(f"Third class total fare: ${third_class_fare:.2f}")

# sabai total into a list
print("\nStep 4: Combine fare totals")
class_fare_totals = [first_class_fare, second_class_fare, third_class_fare]
print(f"Fare totals by class: {class_fare_totals}")

# revenue kati xa
print("\nStep 5: Calculate percentage of fare revenue by class")
fare_percentages = [(fare / total_fare) * 100 for fare in class_fare_totals]

print("\n RESULTS:")
print(f"First Class:  ${first_class_fare:.2f} ({fare_percentages[0]:.2f}%)")
print(f"Second Class: ${second_class_fare:.2f} ({fare_percentages[1]:.2f}%)")
print(f"Third Class:  ${third_class_fare:.2f} ({fare_percentages[2]:.2f}%)")
print(f"Total:        ${total_fare:.2f} (100.00%)")


print("\n QUESTION 2: Passenger Percentage by Age Group")

print("\nStep 1: Creating age_group column")
print("Categories: child (< 18), adult (18-64), senior (>= 65)")

def categorize_age(age):
    if pd.isna(age):
        return None
    elif age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

data['age_group'] = data['Age'].apply(categorize_age)
print("\nSample of age_group column:")
print(data[['Name', 'Age', 'age_group']].head(10))

# passenger ko total number
print("\nStep 2: Calculate total number of passengers (with valid age)")
total_passengers = data['age_group'].notna().sum()
print(f"Total passengers with valid age: {total_passengers}")

# hareko group ko passenger counting
print("\nStep 3: Count passengers in each age group")
age_group_counts = data['age_group'].value_counts()
print(age_group_counts)

# age anusar passenger
print("\nStep 4: Calculate percentage of passengers in each age group")
age_group_percentages = (age_group_counts / total_passengers) * 100

# percent ma display
print("\n RESULTS:")

for age_group in ['child', 'adult', 'senior']:
    if age_group in age_group_counts.index:
        count = age_group_counts[age_group]
        percentage = age_group_percentages[age_group]
        print(f"{age_group.capitalize():8} : {count:4} passengers ({percentage:.2f}%)")
    else:
        print(f"{age_group.capitalize():8} : {0:4} passengers (0.00%)")

print(f"{'Total':8} : {total_passengers:4} passengers (100.00%)")
```
